```python
# Code for extracting maximum flow from Accumulated flow models with possible other delineation (e.g., other DEM)
import arcpy
import numpy  
import csv  
arcpy.SetLogHistory(False)
ShedList = "GRDCSheds_WorldWide" 
PointList = "GRDCPoints_All_Snapped"
expression="DH_ID  > 890"
# make sure there are no lingering layers with locks
f = ["ClippedExtent","ClippedExtentFull", "SnapPour","SnappedAggregated","SnappedPourpoint","InBetween","PourPoly","TestedShed","SelectedPoint","ClippedExtentBuffer","TestedShed_Buffer"]
for x in f:
    arcpy.Delete_management(x)
with arcpy.da.SearchCursor(ShedList, ["DH_ID"], where_clause=expression) as cursor: 
  with arcpy.EnvManager(cellSize = 0.001):
    # run through all sheds
    for row in cursor:   
          # Name and select Shed, including names for when the shed or pour point is not existing
          exp = '"DH_ID" = ' + str(row[0])  
          #print(exp)
          expWrongShed = '"DH_ID" = ' + str(row[0]) + str(' Shed is empty')  
          expWrongPoint = '"DH_ID" = ' + str(row[0]) + str(' Point is empty')  
          arcpy.analysis.Select(ShedList, "TestedShed", exp)
          # add a buffer around the shed to counter for pour points outside the shed (it occurs)  
          arcpy.analysis.Buffer("TestedShed", "TestedShed_Buffer", "0.05 DecimalDegrees", "FULL", "ROUND", "ALL", None, "PLANAR")
          #tester for whether shed is empty
          CountShed = int(arcpy.GetCount_management("TestedShed").getOutput(0))
          if CountShed > 0:    
              # select GRDC weir  
              arcpy.analysis.Select(PointList, "SelectedPoint", exp)
              #tester for whether shed is empty
              CountPoint = int(arcpy.GetCount_management("SelectedPoint").getOutput(0))
              if CountPoint > 0:
                  # run through data sets 
                  for x in range(0,2):
                    # the data input raster, its output generic name and its table
                     if x == 0:
                        density_ras = "WW_World_Clip.tif" 
                        OutBase = "D:/Data/Dropbox/Lactuca/temp/NewWW_"
                        Zonalrast = "ZonalWW"
                        Type = 99
                     elif x == 1:
                         density_ras = "Aqueductkm2Clip.tif" 
                         OutBase = "D:/Data/Dropbox/Lactuca/temp/NewAqueduct_"
                         Zonalrast = "ZonalAque"
                         Type = 99   
                     else:
                         print("This code stops")
                     # Determine extraction point
                     with arcpy.EnvManager(overwriteOutput=True):
                         # following creates a potentially aligned pour point polygon when needed   
                         arcpy.management.Clip(density_ras,"", "ClippedExtent", "TestedShed", "-999", "ClippingGeometry", "NO_MAINTAIN_EXTENT")
                         arcpy.management.Clip(density_ras,"", "ClippedExtentBuffer", "TestedShed_Buffer", "-999", "ClippingGeometry", "NO_MAINTAIN_EXTENT")
                         ClipextentBuffered = arcpy.Describe("ClippedExtentBuffer").extent
                         if arcpy.sa.Raster("ClippedExtent").maximum != None:
                             with arcpy.EnvManager(extent= ClipextentBuffered):
                                 out_raster = arcpy.sa.Con("ClippedExtent", "ClippedExtent", -999, "VALUE >= 0");
                                 out_raster.save("ClippedExtentFull")
                             
                            # snap to pour point, increase search area we allow it be about 4.5km off the known coordinates
                             out_raster = arcpy.sa.SnapPourPoint("SelectedPoint", "ClippedExtentFull", 0.041665, "DH_ID")
                             out_raster.save("SnapPour")
                             out_raster = arcpy.sa.Aggregate("SnapPour", 10, "MEAN", "EXPAND", "DATA") 
                             out_raster.save("SnappedAggregated")
                             out_raster = arcpy.ia.Int("SnappedAggregated"); 
                             out_raster.save("SnappedPourpoint")
                             arcpy.conversion.RasterToPolygon("SnappedPourpoint", "PourPoly", "NO_SIMPLIFY", "Value", "SINGLE_OUTER_PART", None)
                             f = ["ClippedExtent","ClippedExtentFull", "SnapPour","SnappedAggregated","SnappedPourpoint","ClippedExtentBuffer"]
                             for x in f:
                                 arcpy.Delete_management(x)
                             # Extract Data
                             #wextracting the various statistics and joining the values with double check
                             arcpy.sa.ZonalStatisticsAsTable("TestedShed", "OBJECTID", density_ras, "InBetween" ,"DATA", "ALL")
                             arcpy.management.AlterField("InBetween", "Max", "PerPolygonValue", "PolygonMax", "DOUBLE", 8, "NULLABLE", "DO_NOT_CLEAR")
                             arcpy.management.AddField("InBetween", "DH_ID", "LONG", None, None, None, '', "NULLABLE", "NON_REQUIRED", '')
                             arcpy.management.CalculateField("InBetween", "DH_ID", row[0], "PYTHON3", '', "TEXT")
                             arcpy.sa.ZonalStatisticsAsTable("PourPoly", "OBJECTID", density_ras, Zonalrast ,"DATA", "ALL")
                             arcpy.management.AddField(Zonalrast, "DH_ID", "LONG", None, None, None, '', "NULLABLE", "NON_REQUIRED", '')
                             arcpy.management.CalculateField(Zonalrast, "DH_ID", row[0], "PYTHON3", '', "TEXT")
                             arcpy.management.DeleteField(Zonalrast, "COUNT")
                             arcpy.management.DeleteField(Zonalrast, "AREA")
                             f = ["COUNT","AREA", "PerPolygonValue"]
                             for x in f:
                                 arcpy.management.JoinField(Zonalrast, "DH_ID", "InBetween", "DH_ID", (x))
                             arcpy.management.AddField(Zonalrast, "TransferValue","Double", None, None, None,'',"NULLABLE","NON_REQUIRED",'')   
                             if Type == 99:   
                                 arcpy.management.CalculateField(Zonalrast, "TransferValue", "!MAX! / !COUNT!", "PYTHON3", '', "TEXT")
                                 arcpy.management.CalculateField(Zonalrast, "PerPolygonValue", "!PerPolygonValue! / !COUNT!", "PYTHON3", '', "TEXT")   
                             else:
                                 arcpy.management.CalculateField(Zonalrast, "TransferValue", "!MEAN!", "PYTHON3", '', "TEXT")
                             f = ["InBetween","PourPoly"]
                             for x in f:
                                 arcpy.Delete_management(x)
                             # Write Data to cvs file (note, csv's to join in Matlab)
                             table = Zonalrast
                             # create unique name
                             outfile = OutBase + str(row[0])+ ".csv"
                             fields = arcpy.ListFields(table)  
                             field_names = [field.name for field in fields]  
                             with open(outfile,'w') as f:  
                                 w = csv.writer(f , delimiter = ',', lineterminator = '\n')  
                                 #--write all field names to the output file  
                                 w.writerow(field_names)  
                                 for mow in arcpy.SearchCursor(table):  
                                    field_vals = [mow.getValue(field.name) for field in fields]  
                                    w.writerow(field_vals)
                         else: 
                            expWrongShed2 = expWrongShed + str(' for ') + density_ras 
                            print(expWrongShed2 )
              else:
                  print(expWrongPoint)
          else:     
              print(expWrongShed)
          f = ["TestedShed","SelectedPoint","TestedShed_Buffer"]
          for x in f:
              arcpy.Delete_management(x)  
```
